# Data Visualization
Different plots are created to analyze relationships, distributions, and outliers.


```python
###Scatter Plot
##Sepal Length vs Sepal Width

sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    data=iris
)
plt.title("Sepal Length vs Sepal Width")
plt.show()


## in this graph we clearly see that SETOSA SEPAL-WIDTH is more than VERSICOLOR AND VERGINICA but have less SEPAL-LENGTH LESS THAN VERSICOLOR AND VERGINCA
```

```python
##Petal Length vs Petal Width

sns.scatterplot(
    x="petal_length",
    y="petal_width",
    hue="species",
    data=iris
)
plt.title("Petal Length vs Petal Width")
plt.show()
```

```python
##Histogram
##Petal Length Distribution

iris["petal_length"].hist(bins=20)
plt.title("Petal Length Distribution")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
##SEPAL Length Distribution

iris["sepal_length"].hist(bins=20)
plt.title("Sepal Length Distribution")
plt.xlabel("sepal Length")
plt.ylabel("Frequency")
plt.show()
```

```python
##Box Plot
## TO CHECK OUTLIERS

sns.boxplot(
    y="sepal_length",
    data=iris
)
plt.title("Box Plot of Sepal Length")
plt.show()
```

# Final Insights
- The dataset is clean and contains no missing values.
- Scatter plots show clear separation between species.
- Histograms reveal normal distribution for most features.
- Box plots indicate very few outliers.
Overall, the Iris dataset is well-structured and suitable for
machine learning tasks.
